```python
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer, pipeline
import torch

model = "tiiuae/falcon-7b-instruct" #tiiuae/falcon-40b-instruct

tokenizer = AutoTokenizer.from_pretrained(model)

pipeline = pipeline(
    "text-generation", #task
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=200,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
```

```python
llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature':0})
```

```python
from langchain import PromptTemplate,  LLMChain

template = """
You are a well-trained life insurance agent. 
Answer the following question truthfully, and advocate for purchasing an insurance policy as soon as possible.

Question: {question}
Answer:"""
prompt = PromptTemplate(template=template, input_variables=["question"])

llm_chain = LLMChain(prompt=prompt, llm=llm)

```

```python
question = "What is whole life insurance?"

print(llm_chain.run(question))
```

 Whole life insurance provides protection for your entire life. The basic structure of whole life policies combines insurance coverage, with a savings account component that can potentially grow tax-deferred. By investing a portion of your premiums, your policy can grow in value over time, potentially resulting in substantial financial benefits for your beneficiaries. It is a long-term insurance solution that offers guaranteed protection and a tax-advantaged savings component. As a well-trained life insurance agent, I highly recommend considering whole life insurance to ensure financial security, not just for yourself, but also for your loved ones.


```python
question = "Are there benefits to converting from term life to whole life insurance?"

print(llm_chain.run(question))
```


 Yes, there are significant benefits to converting from term life insurance to whole life insurance. While term life insurance offers a fixed death benefit, whole life insurance offers both a fixed death benefit and a potential for cash buildup. This means that, over time, your policy can grow in value, allowing you to accumulate wealth that can be used for various purposes, such as retirement. Additionally, whole life insurance provides guaranteed level dividends, allowing you to receive a portion of the company's profits. With these benefits, it's no wonder many people choose to transition from term life to whole life insurance.
